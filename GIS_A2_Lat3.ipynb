```python
import geopandas as gpd
```

```python
nama='acehutara.zip'
dt=gpd.read_file(nama)
dt
```

```python
dt.plot(column='nama', legend=True,
          legend_kwds={"loc": "center left", "bbox_to_anchor": (1, 0.5)})
```

```python
dt.to_excel('acehutara_data.xlsx', index=False)
print("Data berhasil disimpan ke 'acehutara_data.xlsx'")
```
