```python
import geopandas as gpd
```

```python
dt=gpd.read_file('Aceh_utara.zip')
dt['nama'].value_counts()
dt.to_excel('Acut.xlsx')
```

```python
import pandas as pd
df=pd.read_excel('Acut.xlsx')
df.head()
```

```python
from sklearn.cluster import KMeans
X=df[['luas','penduduk']]
X.head()
```

```python
kmean=KMeans(n_clusters=4,random_state=42)
df['kluster']=kmean.fit_predict(X)
df.head()
```

```python
merger=dt.merge(df,on='nama',how='left')
merger.head()
```
