```python
import pandas as pd
import geopandas as gpd
```

```python
dt=gpd.read_file('AcehUtara.zip')
dt.plot(column='nama')
```

```python
df=pd.read_excel('acehutara_data.xlsx')
df.head(20)
```

```python
dt.head(20)
```

```python
data=df.merge(dt,left_on='nama',right_on='nama')
data.head()
```

```python
data1=data[['nama','luas','penduduk']]
data1.head()
```

```python
data1['kepadatan']=data1['penduduk']/data1['luas']
```

```python
data1.head()
```

```python
wilayahPadat=data1.loc[data1['kepadatan'].idxmax()]
wilayahPadat.head()
```
